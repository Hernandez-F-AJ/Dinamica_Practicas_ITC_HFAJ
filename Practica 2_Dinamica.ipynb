**Problema 1**
<div style="margin-left: 30px;">
    The rotating crank causes the position of point $P$ as a function of time to be $s = 0.4\sin(2\pi t)$m (see figure 2). (a) Determine the velocity and acceleration of $P$ at $t = 0,375$s. (b) What is the maximum magnitude of the velocity of $P$? (c) When the magnitude of the
velocity of $P$ is a maximum, what is the acceleration of $P$?
</div><br>
<div style="text-align: center;">
  <img src="https://github.com/Hernandez-F-AJ/Dinamica_Practicas_ITC_HFAJ/blob/main/imagenes/Practica%202_Dinamica_Fig%202.png" alt="image" width="300"><br>

Tal que $ v(t) = \frac{d}{dt}s(t)$ y $a(t) = \frac{d}{dt}v(t)$ podemos integrar a $s$ que esta dada en funcion de $t$ &there4;
<div style="margin-left: 160px;">
    $v(t)  = s'= \frac{d}{dt} s(t)$<br>
    $v(t) = \frac{d}{dt}0.4\sin(2\pi t)$<br>
    $v(t) = 0.4\times(2\pi)\cos(2\pi t)=$ <span style="background-color: yellow;"> $0.8\pi\cos(2\pi t)$ </span><br><br>
    $a(t) = v'=\frac{d}{dt}v(t) = \frac{d^2}{dt^2}s(t)$<br>
    $a(t) = \frac{d}{dt}0.8\pi\cos(2\pi t)$<br>
    $a(t) = 0.8\pi\times -2\pi\sin(2\pi t)=$<span style="background-color: yellow;"> $-1.6\pi^2 \sin(2\pi t)$ </span><br>    
</div><br>
a)<span style="padding-left:10px"></span>$t= 0.375$s.
<div style="margin-left: 160px;">
    $v(0.375) = 0.8\pi\cos(2\pi \times 0.375)=$<span style="background-color: yellow;"> $-1.777153175\frac{m}{s}$</span><br><br>
    $a(0.375) = -1.6\pi^2 \sin(2\pi \times 0.375)=$<span style="background-color: yellow;">$-11.16618272\frac{m}{s^2}$<br>
</div></span><br>
b)<span style="padding-left:10px"></span>Magnitud máxima de la velocidad de $P$. Para calcular el valor máximo podemos hacer dos aproximaciones. Dado que la velocidad está en función de una onda cosenoidal en $t$, sabemos que el máximo de $\cos(2\pi t)$ es $1$. Esta onda se ve afectada por el factor de escala $0.8\pi$, por lo tanto:
<div style="margin-left: 160px;">
    $v_{max}(t):= v(t) \quad \text{cuando} \quad \cos(2\pi t) = 1$<br>    
</div><br>
De lo anterior, se obtiene:
<div style="margin-left: 160px;">
   $\left| v_{max}(t) \right|= 0.8 \pi \times 1=$<span style="background-color: yellow;"> $2.513274123\frac{m}{s}$ </span>    
</div><br>
También podemos usar el criterio de la segunda derivada:
<div style="margin-left: 160px;">
    $v'(t)'= 0.8\pi\cos(2\pi t)\frac{d}{dt}= -1.6\pi^2 \sin(2\pi t)$<br>
    $v''(t)= -1.6\pi^2 \sin(2\pi t)\frac{d}{dt}=-3.2\pi^3 \cos(2\pi t)$<br><br>
</div>
Encontramos el valor critico:
<div style="margin-left: 160px;">
    $v'(t)'= -1.6\pi^2 \sin(2\pi t) = 0 \quad \Rightarrow \quad \sin(2\pi t) = 0 \quad \Rightarrow \quad t = 0$
</div><br>
Sustituyendo en $v''(t)$, se obtiene:
<div style="margin-left: 160px;">
    $v''(t)= -3.2\pi^3 \cos(2\pi 0)= -99.22008538$<br>
</div><br>
Ya que al evaluar $t$ de $v'(t)$ se obtiene un resultado $<0$, entonces tenemos un máximo para $t=0$, por lo que:
<div style="margin-left: 160px;">
    $v(0)= 0.8\pi\cos(2\pi 0)=$<span style="background-color: yellow;"> $2.513274123\frac{m}{s}$ </span>
</div><br>
    
c)<span style="padding-left:10px"></span> Cuando la magnitud de la velocidad de $P$ es máxima, ¿Cuál es la aceleración de p?
Cuando $v(t)$ es un máximo, se tiene que $t =0$, entonces evaluemos $a(t)$ para $t = 0$:
<div style="margin-left: 160px;">
    $a(0) = -1.6\pi^2 \sin(2\pi \times 0)=$<span style="background-color: yellow;"> $0\frac{m}F{s^2}$ </span>
</div><br><br>

Ahora usaremos herramientas de programación para realizar los calculos y observar si estos valores son similares con los previamente calculados:

```python
# Problema
t = sp.symbols('t', real=True)
s = 0.4 * sp.sin(2 * sp.pi * t)
v = sp.diff(s, t)
a = sp.diff(v, t)

# a) Evaluación en t = 0.375 s
t_eval = 0.375
v_a = v.subs(t, t_eval).evalf()
a_a = a.subs(t, t_eval).evalf()


print("a) En t = 0.375 s:")
print("v(0.375) =", v_a, "m/s")
print("a(0.375) =", a_a, "m/s²")


# b) Máxima magnitud de la velocidad

# Método 1: amplitud del coseno
v_max_directo = abs(0.8 * np.pi)

# Método 2: criterio de derivadas
criticos = sp.solve(sp.Eq(a, 0), t)  
valores_v = [v.subs(t, c).evalf() for c in criticos]
idx_max = np.argmax([abs(val) for val in valores_v])
t_max_v = criticos[idx_max]
v_max_derivadas = valores_v[idx_max]

print("\n b) Máxima magnitud de la velocidad:")
print("Método directo:", v_max_directo, "m/s")
print("Método derivadas: |v| =", abs(v_max_derivadas), "m/s en t =", t_max_v)


# c) Aceleración cuando |v| es máxima
a_max_v = a.subs(t, t_max_v).evalf()

print("\n c) Aceleración cuando |v| es máxima:")
print("a(t) en t =", t_max_v, "es", a_max_v, "m/s²")
```

**Problema 2**
For the mechanism of problem 1 draw graphs of the position $s$, velocity $v$, and
acceleration a of point $P$ as functions of time for $0 ≤ t ≤ 2$s. Using your graphs, confirm
that the slope of the graph s is zero at times for which v is zero, and the slope of the graph
v is zero at times for which a is zero.

```python
# Tiempo de 0 a 2 segundos
t = np.linspace(0, 2, 10000)

# Funciones
s = 0.4 * np.sin(2 * np.pi * t)
v = 0.8 * np.pi * np.cos(2 * np.pi * t)
a = -1.6 * (np.pi**2) * np.sin(2 * np.pi * t)

# Tolerancia
tol = 1e-2   # más laxo para la derivada numérica

# Ceros de v(t)
v_zero = np.where(np.abs(v) < tol)[0]

# Ceros de a(t)
a_zero = np.where(np.abs(a) < tol)[0]

# Derivada numérica de v(t) para detectar máximos/mínimos
dv_dt = np.gradient(v, t)

# Tolerancia
idx_dvdt_zero_tol = np.where(np.abs(dv_dt) < tol)[0]

# Subgráficas
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# s(t)
axs[0].plot(t, s, label='s(t)', color='blue')
axs[0].plot(t[v_zero], s[v_zero], 'o', color='orange', label='Pendiente de s = 0')
axs[0].set_ylabel('Posición (m)')
axs[0].legend()
axs[0].grid(True)

# v(t)
axs[1].plot(t, v, label='v(t)', color='green')
axs[1].plot(t[v_zero], v[v_zero], 'o', color='orange', label='v = 0')
axs[1].plot(t[idx_dvdt_zero_tol], v[idx_dvdt_zero_tol], 'o', color='purple', label='Pendiente de v = 0')
axs[1].set_ylabel('Velocidad (m/s)')
axs[1].legend()
axs[1].grid(True)

# a(t)
axs[2].plot(t, a, label='a(t)', color='red')
axs[2].plot(t[a_zero], a[a_zero], 'o', color='purple', label='a = 0')
axs[2].set_ylabel('Aceleración (m/s²)')
axs[2].set_xlabel('Tiempo (s)')
axs[2].legend()
axs[2].grid(True)

plt.suptitle('Gráficas con ceros de v(t), a(t) y pendientes de v', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
