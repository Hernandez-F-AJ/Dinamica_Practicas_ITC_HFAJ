a)<span style="padding-left:10px"></span>$ \int_{0}^{1} (3x^2 + 2) dx $

<div style="margin-left: 160px;">
    $\int 3x^2 dx=x^3$<span style="padding-left:10px"></span>;<span style="padding-right:10px"></span> $\int 2 dx=2x$<br>
    $\left[x^3 + 2x + c\right]_0^1 =(1^3 +2(1))-(0^3 +2(0))$<br>
    <span style="background-color: yellow;">$ \int_{0}^{1} (3x^2 + 2) = 3$</span>
</div><br>

b)<span style="padding-left:10px"></span>$ \int_{0}^{1} (x^5 -5x +6) dx $

<div style="margin-left: 160px;">
    $\int x^5 dx= \frac{x^6}{6} $<span style="padding-left:10px"></span>;<span style="padding-right:10px"></span> $\int -5x dx= -\frac{5x^2}{2}$<span style="padding-left:10px"></span>;<span style="padding-right:10px"></span> $\int 6 dx= 6x$<br>
    $\left[ \frac{x^6}{6} -\frac{5x^2}{2} + 6x \right]_0^1 =(\frac{1^6}{6} -\frac{5(1)^2}{2} + 6(1))-(\frac{0^6}{6} -\frac{5(0)^2}{2} + 6(0))$<br>
    <span style="background-color: yellow;"> $\int_{0}^{1} (x^5 -5x +6) dx = \frac{11}{3} $ </span>
</div><br>

c)<span style="padding-left:10px"></span>$ \int_{-\pi/2}^{\pi/2} \cos(x) dx $

<div style="margin-left: 160px;">
    $\int \cos(x) dx= \sin(x)$<br>
    $\left[ \sin(x)\right]_{-\pi/2}^{\pi/2} =(\sin(\pi/2))-(\sin(-\pi/2))$<br>
<span style="background-color: yellow;"> $ \int_{-\pi/2}^{\pi/2} \cos(x) dx = 2$ </span>
</div><br>

<div style="margin-left: 160px;">
    $\int \sin^2(x) dx= -\frac{\cos(x)\sin(x)}{2} + \frac{1}{2} \int 1 dx = -\frac{\sin(2x)-2x}{4}$<br>
    $\left [-\frac{\sin(2x)-2x}{4}\right]_{0}^{\pi/2} =(-\frac{\sin(2\times \pi/2)-2(\pi/2)}{4}))-(-\frac{\sin(2(0))-2(0)}{4})$<br>

f)<span style="padding-left:10px"></span>Comprueba tus resultados con Matlab.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
from matplotlib.ticker import MultipleLocator
# a)
def f_a(x):
    return 3*x**2 + 2
res_a, err_a = quad(f_a, 0, 1)

# b)
def f_b(x):
    return x**5 - 5*x + 6
res_b, err_b = quad(f_b, 0, 1)

# c)
def f_c(x):
    return np.cos(x)
res_c, err_c = quad(f_c, -np.pi/2, np.pi/2)

# d)
def f_d(x):
    return np.sin(x)**2
res_d, err_d = quad(f_d, 0, np.pi/2)

# e)
def f_e(x):
    return np.exp(-x)
res_e, err_e = quad(f_e, 0, 1)

# Resultados
print(f"a) Integral = {res_a:.6f}")
print(f"b) Integral = {res_b:.6f}")
print(f"c) Integral = {res_c:.6f}")
print(f"d) Integral = {res_d:.6f}")
print(f"e) Integral = {res_e:.6f}")
print()
```
